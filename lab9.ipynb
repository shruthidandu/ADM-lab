```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(42)
X = np.linspace(0, 10, 50).reshape(-1, 1)
y = 2 + 3 * X - 0.5 * X**2 + np.random.normal(0, 1, X.shape)
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
y_pred = model.predict(X_poly)
```

```python
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
plt.scatter(X, y, color='blue', label="Original Data")
plt.plot(X, y_pred, color='red', label="Polynomial Regression Line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python

degree = 3
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(X)
```

```python

model = LinearRegression()
model.fit(X_poly, y)
```

```python

y_pred = model.predict(X_poly)
```

```python
from sklearn.metrics import mean_squared_error, r2_score # Import r2_score

mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python

X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
y_range_pred = model.predict(X_range_poly)
```

```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X_range, y_range_pred, color='red', label=f'Polynomial Degree {degree}')
plt.title('Polynomial Regression Curve')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```
